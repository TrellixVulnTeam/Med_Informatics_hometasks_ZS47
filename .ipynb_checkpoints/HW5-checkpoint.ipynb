### Opening file and dividing target and predictors


```python
import pandas as pd
import numpy as np

#read file
df = pd.read_excel('C:\\Amanturs new\\study\\1sem\\1half\\med informatics\\HW5\\Data\\data.xlsx')
# df.head(10)
# df = df.replace({'diagnosis': {'M': 1, 'B': 0}})
Y = df['diagnosis']
Y = pd.Series(np.where(Y.values == 'M', 1, 0), Y.index)
X = df.drop(['diagnosis', 'id'], 1)
Y.head(10)
```

### Adding model and metrics

```python
from sklearn import linear_model
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

model = linear_model.LogisticRegression(max_iter=3000,solver='lbfgs')
```

### Task 5

```python
def find_feature(df, iterations, max_score=0, f_list=[]):
    if iterations != 0:
        Xnf = df
        for f in Xnf.columns:
            model.fit(Xnf[f_list+[f]], Y)
            Ypred = model.predict(Xnf[f_list+[f]])
            score = accuracy_score(Y, Ypred)
            if max_score < score:
                max_score = score
                best_f = f
        f_list.append(best_f)
        find_feature(X, iterations-1, max_score, f_list)
    else:
        print(f_list, max_score)

                                  
find_feature(df=X, iterations=3)
```

### Task 6

```python
def find_feature_rec(df, iterations, max_rec=0, f_list=[]):
    if iterations != 0:
        Xnf = df
        for f in Xnf.columns:
            model.fit(Xnf[f_list+[f]], Y)
            Ypred = model.predict(Xnf[f_list+[f]])
            rec = recall_score(Y, Ypred)
            if max_rec < rec:
                max_rec = rec
                best_f = f
        f_list.append(best_f)
        find_feature_acc(X, iterations-1, max_rec, f_list)
    else:
        print(f_list, max_rec)

                                  
find_feature_rec(df=X, iterations=2)
```

### Task 7

```python
from sklearn import tree
from matplotlib import pyplot as plt

decision_tree = tree.DecisionTreeClassifier(max_leaf_nodes=3)
decision_tree.fit(X,Y)
fig = plt.figure(figsize=(15,10))
_ = tree.plot_tree(decision_tree, feature_names=list(X.columns), class_names=['B','M'], rounded = True, fontsize = 12, filled=True)
Ypred = decision_tree.predict(X)
score = accuracy_score(Y, Ypred)
score
```
